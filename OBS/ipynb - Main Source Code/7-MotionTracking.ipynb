```python
import cv2

cap = cv2.VideoCapture(0)
_, f1 = cap.read()
_, f2 = cap.read()

while True:
    diff = cv2.absdiff(f1, f2)
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    _, th = cv2.threshold(gray, 25, 255, cv2.THRESH_BINARY)

    cv2.imshow("Motion", th)

    f1 = f2
    _, f2 = cap.read()

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
