```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("person.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Define skin color range in HSV
lower_skin = np.array([0, 48, 80])
upper_skin = np.array([20, 255, 255])

# Create mask
mask = cv2.inRange(hsv, lower_skin, upper_skin)

# Apply mask
skin = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

# Show results
plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(mask, cmap='gray')
plt.title("Skin Mask")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(skin)
plt.title("Detected Skin")
plt.axis("off")

plt.show()
```
